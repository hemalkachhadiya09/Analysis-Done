Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select 
only numbers divisible by 7 and 17

```python
sr=pd.Series(np.arange(1,1001))
print(sr[(sr%7==0)&(sr%17==0)])
```

Problem Statement 4: Sylphia has a dataset of various cereals sold in the supermarket.


```python
!wget https://www.dropbox.com/s/idnul34dfo5cnke/cereal.csv
```

```python
data = pd.read_csv('cereal.csv')
data.head()
```

```python
rating = data.groupby('mfr')['rating'].mean()
rating
```

```python
plt.bar(rating.index, rating.values)
plt.show()
```

```python
plt.barh(rating.index,rating.values)
plt.show()
```

```python
plt.barh(rating.index,rating.values)
plt.xlim(0,100)
plt.show()
```

```python
plt.style.use('seaborn')
plt.barh(rating.index, rating.values)
plt.xlim(0,100)
plt.show()
```
